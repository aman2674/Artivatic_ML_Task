## Problem Statement
- The Bank Indessa has not done well in the last 3 quarters. Their NPAs (Non Performing Assets) have reached all time high. It is starting to lose the confidence of its investors. As a result, it’s stock has fallen by 20% in the previous quarter alone.
- After careful analysis, it was found that the majority of NPA was contributed by loan defaulters. With the messy data collected over all the years, this bank has decided to use machine learning to figure out a way to find these defaulters and devise a plan to reduce them.
- This bank uses a pool of investors to sanction theirloans.For example: If customer has applied for loan of $20000,  along with the bank, the investors perform due diligence on the requested loan application. Keep this in mind while understanding data.

```python
#Replacing missing values with new label....
def replace_missing_value(df,categorical_with_nan):
    data = df.copy()
    data[categorical_with_nan] = data[categorical_with_nan].fillna("Missing")
    return data
df_train = replace_missing_value(df_train,categorical_with_nan)
df_test = replace_missing_value(df_test,categorical_with_nan)
```

### Numerical Variable

```python
numerical_with_nan = [feature for feature in numerical_feature if df_train[feature].isnull().sum()>0]
numerical_with_nan
```

```python
#Replacing the numerical missing values....
def replace_with_median(df,numerical_with_nan):
    for feature in numerical_with_nan:
        #replace missing values with  median because there are outliers present 
        median_value = df[feature].median()
        #create new feature to capture more info of nan value
        df[feature].fillna(median_value,inplace=True)
    return df
df_train = replace_with_median(df_train,numerical_with_nan)
df_test = replace_with_median(df_test,numerical_with_nan)
```
